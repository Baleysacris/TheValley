# Modificaciones sobre dataframes origen
Modificaciones a dataframes originales para prepararlos para EDA Visual.




# 01 Import

```
import pandas as pd
import numpy as np
```

# 02 Load Data

```
# df_covid

# url de archivo raw de casos confirmados
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

# carga el archivo en un dataframe df_covid
df_covid = pd.read_csv(url)

df_covid.head()
```

# 03 Transformaciones dataset Covid-19
Queremos hacer un análisis visual que incluya a España, USA, Francia, Reino Unido, Francia, Italia, Portugal y Alemania. Así que como paso previo, tendremos que crear un nuevo dataframe df_covid_clean sin el resto de países.

Nota: Algunos de los países deseados tienen sus datos desagregados en varias provincias. Agrúpalos en una sola entrada. Es decir, no queremos en este análisis tener una fila que representa a Nueva York y otra que representa a California, sino una única que representa a USA donde se incluyen ambas. De hecho, puedes eliminar la columna Provincia del dataframe df_covid_clean.

```
# Países totales
df_covid['Country/Region'].unique()
```

```
# Lista con los paises deseados
lista_paises = ['China', 
                'France', 'Germany', 'Italy', 
                'Portugal', 'Spain', 'US', 'United Kingdom']
```

```
# Filtra el dataframe para crear una nuevo
filtro_paises = ((df_covid['Country/Region'].isin(lista_paises))&
                 (df_covid['Province/State'].isna()))
df_covid_clean_no_china = df_covid.loc[filtro_paises, :]
df_covid_clean_no_china
```

```
# Paises sin China en orden
cols = list(df_covid_clean_no_china['Country/Region'])
cols
```

```
# Datos para paises no China
datos_sin_china = df_covid_clean_no_china.iloc[:,4:].T
datos_sin_china.columns = ['UK' if pais=='United Kingdom' else pais for pais in cols]
datos_sin_china.index = pd.to_datetime(df_covid_clean_no_china.columns[4:],format='%m/%d/%y')
datos_sin_china
```

```
# Datos China
df_covid_china = df_covid.loc[df_covid['Country/Region']=='China',:]
df_covid_china = df_covid_china.sum().reset_index().iloc[4:,:]
df_covid_china.columns = ['Fecha', 'Confirmados']
df_covid_china
```

```
#  Todos los datos
df_covid_clean = datos_sin_china.copy(deep=True)
df_covid_clean['China'] = df_covid_china['Confirmados'].values
df_covid_clean
```

```
# Datos Relativizados
df_covid_clean['France_REL'] = df_covid_clean['France']/67060000*100000
df_covid_clean['Germany_REL'] = df_covid_clean['Germany']/83020000*100000
df_covid_clean['Italy_REL'] = df_covid_clean['Italy']/60360000*100000
df_covid_clean['Portugal_REL'] = df_covid_clean['Portugal']/10280000*100000
df_covid_clean['Spain_REL'] = df_covid_clean['Spain']/49940000*100000
df_covid_clean['US_REL'] = df_covid_clean['US']/328200000*100000
df_covid_clean['UK_REL'] = df_covid_clean['UK']/66650000*100000
df_covid_clean['China_REL'] = df_covid_clean['China']/1398000000*100000
```

```
# Nuevos casos ultimos 7 días
for col in df_covid_clean.columns:
  df_covid_clean[col+'_ult7'] = df_covid_clean[col] - df_covid_clean[col].shift(7)
```

```
df_covid_clean.insert(0, 'Date', df_covid_clean.index)
df_covid_clean.reset_index(drop=True, inplace=True)
df_covid_clean.head()
```
